Preparing the data

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

### **Training the Model:**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_
# Plotting for the test data
plt.scatter(X, y,color="red")
plt.plot(X, line);
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')
plt.show()
```


### Predicting the result

```python
# Testing data - In Hours
print(X_test)
```

Model Prediction

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

### Comparing Actual values with Predicted values :

```python
dataset = pd.DataFrame({'Actual Score': y_test, 'Predicted Score': y_pred})  
dataset
```

### **Predicting the score of the student studied for 9.25 hours/day**

```python
print('Score of student who studied for 9.25 hours per day is ', regressor.predict([[9.25]]))
```

### Evaluating the Model :

```python
#Checking the efficiency of model
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-Square:', metrics.r2_score(y_test, y_pred))
```
